### **Visualize Data**


```python
import pandas as pd

# Load the dataset from the CSV file
df = pd.read_csv('park.csv')

# select the top 25 neighborhoods and sort them alphabetically by neighborhood name, and plot the result as a bar chart
df['neighborhood'].value_counts().head(25).sort_index().plot(kind='bar', title="Top 25 Neighborhoods by Park Count")

# Sort the original DataFrame by the 'neighborhood' column alphabetically 
df_sorted = df.sort_values(by='neighborhood', ascending=True)


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset, count the number of parks in each neighborhood
df = pd.read_csv('park.csv')

top_parks = df['neighborhood'].value_counts().head(25).sort_values(ascending=False)

# Create the plot
plt.figure(figsize=(10, 6))  # Set figure size
top_parks.plot(kind='bar', color='purple', edgecolor='black', width=0.8)

# Adding labels and title
plt.title("Top 25 Neighborhoods by Park Count", fontsize=16)
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Number of Parks', fontsize=12)

# Adding gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Displaying the plot
plt.tight_layout()  # Adjust layout for better fit
plt.show()

```
