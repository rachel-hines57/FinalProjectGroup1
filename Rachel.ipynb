## Submetric

The submetric I chose that can be used to determine which neighborhood is the most outdoor-friendly is public art displays. My data set lists public art displays in different neighborhoods of Pittsburgh. I used the data to determine which neighboorhood has the most public art displays, which makes the neighborhood more enjoyable to be outside in.

## Import Pandas and Read in Data Set

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
art = pd.read_csv("PublicArt.csv")
art.head()
```

## Make a Bar Chart
This bar chart uses just the neighboorhood column of the data to show the count of public art displays in each neighborhood, which is my variable of interest.

```python
art["neighborhood"].value_counts().sort_index().plot(kind = "bar", rot = 90, figsize = (10,5), xlabel = "Neighborhood", ylabel = "Number of Public Art Displays")
```

## Sort the Bar Graph from Greatest to Least
I got feedback on the first graph I made that I should sort the bars from greatest to least number of public art displays. This way, it is easy for viewers to see which neighborhood has the most. 

```python
art["neighborhood"].value_counts().sort_values(ascending=False).plot(kind = "bar", rot=90,figsize = (10,5), xlabel = "Neighborhood", ylabel= "Number of Public Art Displays")
```

## Conclusion
Based on the bar chart, the Central Business District has the greatest amount of public art displays, with the other nieghborhoods not coming close to its ount of over 25. This means that Central Business District wins in terms of the public art aspect of having an outdoor-friendly neighborhood.
